```python
import gensim.downloader as api

# Load the pretrained Word2Vec model
model = api.load('word2vec-google-news-300')

# Choose 5 words
words = ["doctor", "apple", "university", "cricket", "car"]

# Display top 5 similar words for each
for word in words:
    print(f"\nTop similar words to '{word}':")
    try:
        similar_words = model.most_similar(word, topn=5)
        for sim_word, score in similar_words:
            print(f"  {sim_word} ({score:.4f})")
    except KeyError:
        print(f"  Word '{word}' not found in the vocabulary.")
```

```python
examples = [
    ("Paris", "France", "Italy"),     # Expected ≈ Rome
    ("Doctor", "Man", "Woman"),       # Expected ≈ Nurse
    ("Apple", "Fruit", "Computer"),   # Expected ≈ Mac
]

print("\nVector arithmetic results:\n")

for a, b, c in examples:
    try:
        result = model.most_similar(positive=[a, c], negative=[b], topn=1)
        predicted_word, similarity = result[0]
        print(f"{a} - {b} + {c} ≈ {predicted_word} ({similarity:.4f})")
    except KeyError as e:
        print(f"  One of the words not found in vocabulary: {e}")
```
